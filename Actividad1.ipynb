```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ejercicio 1.
def f1(x):
    return np.cos(x)-x*np.exp(x)
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(0, 1, 100)
```

```python
# Graficar ambas funciones.
plt.plot(x, [f1(i) for i in x],color="red")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(0, 1)
plt.ylim(-1, 1.5)
plt.title(r'f(x)= $cos(x)-xe^x$')
plt.grid()
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.show()
```

```python
# Ejercicio 2
def f2(x):
    return 10**x - x - 4
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(0, 1, 70)
```

```python
# Graficar ambas funciones.
plt.plot(x, [f2(i) for i in x],color="red")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(0, 1)
plt.ylim(-1, 1)
plt.title(r'f(x)= $10^x-x -4$')

plt.grid()
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.show()
```

```python
X=np.linspace(0.66,0.67,11)
for x in X:
  print(x,f2(x))
```

```python
# Ejercicio 3.
def f3(x):
    return x**3+4*x**2-10
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(1, 2, 5)
```

```python
# Graficar ambas funciones.
plt.plot(x, [f3(i) for i in x],color="red")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(1, 2)
plt.ylim(-1, 1)
plt.title(r'f(x)= $x^3 +4x^2 - 4$')
plt.grid()
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.show()
```
